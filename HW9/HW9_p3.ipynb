\begin{aligned}
\tilde v_{1} &=\begin{bmatrix}1\\1\\1\\1\end{bmatrix},\qquad 
\tilde v_{2} &=\begin{bmatrix}1\\1\\1\\0\end{bmatrix},\qquad
\tilde v_{3} &=\begin{bmatrix}1\\1\\0\\0\end{bmatrix}
\end{aligned}

$\textbf{Step 1: } u_{1}= \tilde v_{1},\;$
$
\lVert u_{1}\rVert=\sqrt{1^{2}+1^{2}+1^{2}+1^{2}} = 2,
\qquad
q_{1}= \dfrac{u_{1}}{\lVert u_{1}\rVert}
      =\frac12\!\begin{bmatrix}1\\1\\1\\1\end{bmatrix}
$

$
\textbf{Step 2: } 
\text{proj}_{q_{1}}\!\bigl(\tilde v_{2}\bigr)
     =(\tilde v_{2}\!\cdot\! q_{1})\,q_{1}
     =\Bigl(\frac12+\frac12+\frac12+0\Bigr)q_{1}
     =\frac32\,q_{1}
     =\frac34\!\begin{bmatrix}1\\1\\1\\1\end{bmatrix}
$

$
u_{2}= \tilde v_{2}-\text{proj}_{q_{1}}(\tilde v_{2})
     =\begin{bmatrix}\tfrac14\\\tfrac14\\\tfrac14\\-\tfrac34\end{bmatrix},
\qquad
\lVert u_{2}\rVert
    =\sqrt{3\!\left(\tfrac14\right)^{\!2}
            +\left(-\tfrac34\right)^{\!2}}
    =\sqrt{\tfrac34}
    =\frac{\sqrt3}{2}
$


$
q_{2}= \dfrac{u_{2}}{\lVert u_{2}\rVert}
     =\frac{2}{\sqrt3}\!
      \begin{bmatrix}\tfrac14\\\tfrac14\\\tfrac14\\-\tfrac34\end{bmatrix}
     =\begin{bmatrix}
        \dfrac1{2\sqrt3}\\[4pt]
        \dfrac1{2\sqrt3}\\[4pt]
        \dfrac1{2\sqrt3}\\[4pt]
        -\dfrac{3}{2\sqrt3}
       \end{bmatrix}
$


$
\textbf{Step 3: } 
\text{proj}_{q_{1}}\!\bigl(\tilde v_{3}\bigr)
     =(\tilde v_{3}\!\cdot\! q_{1})\,q_{1}
     =(1+1)\!\left(\frac12\right)q_{1}=1\,q_{1}
     =\frac12\!\begin{bmatrix}1\\1\\1\\1\end{bmatrix}
$


$
\text{proj}_{q_{2}}\!\bigl(\tilde v_{3}\bigr)
     =(\tilde v_{3}\!\cdot\! q_{2})\,q_{2}
     =\bigl( \tfrac1{2\sqrt3} + \tfrac1{2\sqrt3} \bigr) q_{2}
     =\frac1{\sqrt3}\,q_{2}
     =\begin{bmatrix}\tfrac16\\\tfrac16\\\tfrac16\\-\tfrac12\end{bmatrix}
$


$
u_{3}= \tilde v_{3}-\text{proj}_{q_{1}}(\tilde v_{3})
               -\text{proj}_{q_{2}}(\tilde v_{3})
     =\begin{bmatrix}\tfrac13\\\tfrac13\\-\tfrac23\\0\end{bmatrix},
\qquad
\lVert u_{3}\rVert
     =\sqrt{\left(\tfrac13\right)^{2}
            +\left(\tfrac13\right)^{2}
            +\left(-\tfrac23\right)^{2}}
     =\sqrt{\tfrac23}
$

$
q_{3}= \dfrac{u_{3}}{\lVert u_{3}\rVert}
     =\sqrt{\tfrac32}\!
       \begin{bmatrix}\tfrac13\\\tfrac13\\-\tfrac23\\0\end{bmatrix}
     =\begin{bmatrix}
        \dfrac1{\sqrt6}\\[4pt]
        \dfrac1{\sqrt6}\\[4pt]
        -\dfrac2{\sqrt6}\\[4pt]
        0
       \end{bmatrix}
$

$
\boxed{
\bigl\{\,q_{1},\,q_{2},\,q_{3}\bigr\}
   =\left\{
      \frac12\!\begin{bmatrix}1\\1\\1\\1\end{bmatrix},
      \begin{bmatrix}\dfrac1{2\sqrt3}\\ \dfrac1{2\sqrt3}\\ \dfrac1{2\sqrt3}\\ -\dfrac{3}{2\sqrt3}\end{bmatrix},
      \begin{bmatrix}\dfrac1{\sqrt6}\\ \dfrac1{\sqrt6}\\ -\dfrac2{\sqrt6}\\ 0\end{bmatrix}
     \right\}
\text{ is an orthonormal basis for }W.
}
