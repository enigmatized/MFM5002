```python
import numpy as np
import matplotlib.pyplot as plt

# Exact value
I_exact = 0.5
function_to_integrate = lambda x: 2 * x**3

def monte_carlo_integration(function_to_integrate, I_exact):
    # Store results in a list: (i, I_est, absolute_error)
    res = []
    for i in range(1, 7):
        N = 10**i
        x = np.random.rand(N)
        f_values = function_to_integrate(x)
        I_est = np.mean(f_values)
        res.append((N, I_est, abs(I_est - I_exact)))

    # Separate out for plotting convenience
    Ns = [item[0] for item in res]
    I_ests = [item[1] for item in res]
    errors = [item[2] for item in res]

    # --- PLOT 1: Monte Carlo estimate vs number of samples ---
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(Ns, I_ests, marker='o', linestyle='--', label='MC Estimate')
    plt.axhline(I_exact, color='r', linestyle='-', label='Exact = 0.5')
    plt.xscale('log')  # Log scale for N on x-axis
    plt.xlabel('Number of samples (N)')
    plt.ylabel('Monte Carlo Estimate')
    plt.title('Estimate vs. Number of Samples')
    plt.legend()

    # --- PLOT 2: Absolute error vs number of samples ---
    plt.subplot(1, 2, 2)
    plt.plot(Ns, errors, marker='o', linestyle='--', label='Absolute Error')
    plt.xscale('log')  # Log scale for N
    plt.yscale('log')  # Log scale for error (helps show error decay)
    plt.xlabel('Number of samples (N)')
    plt.ylabel('Absolute Error')
    plt.title('Error vs. Number of Samples')
    plt.legend()

    plt.tight_layout()
    plt.show()
monte_carlo_integration(function_to_integrate, I_exact)
```
