$f(x) = \sum_{n=0}^{\infty} \frac{f^{(n)}(x₀)}{n!}(x - x₀)^n$

$f(x, y) ≈ f(x₀, y₀) + \frac{∂f}{∂x}(x₀, y₀)(x - x₀) + \frac{∂f}{∂y}(x₀, y₀)(y - y₀)
          + \frac{1}{2!} \left[ \frac{∂²f}{∂x²}(x₀, y₀)(x - x₀)² + 2\frac{∂²f}{∂x∂y}(x₀, y₀)(x - x₀)(y - y₀) + \frac{∂²f}{∂y²}(x₀, y₀)(y - y₀)² \right]
          + ...$


In Summation notation

$ f(x,y) = \sum_{n=0}^{\infty} \sum_{m=0}^{\infty} \frac{1}{n!m!}\frac{\partial^{(n+m)}f}{\partial x^n \partial y^m}(x_0,y_0) (x-x_0)^n (y-y_0)^m$

So we are given

$f(x, y) = e^{x + y²}$

And luckily we are also given $(x₀, y₀) = (0, 0)$ (Luckily because we know we will get a lot of $e^0$)


So now is the mechnaical part.

#### First calculate partial Derivatives

- $\frac{∂f}{∂x} = e^{x + y²}$

- $ \frac{∂f}{∂y} = 2ye^{x + y²}$

- $\frac{∂²f}{∂x²} = e^{x + y²}$

- $ \frac{∂²f}{∂y²} = 2e^{x + y²} + 4y²e^{x + y²}$

- $\frac{∂²f}{∂x∂y} = 2ye^{x + y²}$


##### Next plug in (0,0)

- $f(0, 0) = e⁰ = 1$

- $ \frac{∂f}{∂x}(0, 0) = e⁰ = 1$

- $ \frac{∂f}{∂y}(0, 0) = 2(0)e⁰ = 0$

- $ \frac{∂²f}{∂x²}(0, 0) = e⁰ = 1$

- $ \frac{∂²f}{∂y²}(0, 0) = 2e⁰ + 4(0)²e⁰ = 2$ 

- $ \frac{∂²f}{∂x∂y}(0, 0) = 2(0)e⁰ = 0 $


##### Putting it all together

$f(x, y) ≈ 1 + 1(x) + 0(y) + \frac{1}{2} [1(x)² + 2(0)(x)(y) + 2(y)²]$

$ f(x, y) \approx 1 + x + \frac{1}{2}x^2 + y^2$

## Delta–Gamma–Theta Approximations

\[
\begin{aligned}
&\text{Let }V(S_t)\text{ be the option price at time }t\text{, where }S_t\text{ is the underlying stock price at time }t.\\
&\text{Let }h\text{ be a small increment in time, and let }\epsilon = S_{t+h} - S_t\text{ be the change in the stock price over }(t,t+h].\\[6pt]
&\text{We define }\Delta = \frac{\partial V}{\partial S},\quad
 \Gamma = \frac{\partial^2 V}{\partial S^2},\quad
 \theta = \frac{\partial V}{\partial t}.\\[6pt]
&\text{Then the Taylor‐series expansions yield the following approximations:}\\
&\quad \text{(1) Delta approximation:}\\
&\quad\quad V(S_{t+h}) \approx V(S_t) \;+\; \Delta \,\epsilon,\\[6pt]
&\quad \text{(2) Delta–Gamma approximation:}\\
&\quad\quad V(S_{t+h}) \approx V(S_t) \;+\; \Delta \,\epsilon \;+\; \tfrac{1}{2}\,\Gamma\,\epsilon^2,\\[6pt]
&\quad \text{(3) Delta–Theta approximation:}\\
&\quad\quad V(S_{t+h}) \approx V(S_t) \;+\; \Delta \,\epsilon \;+\; h\,\theta,\\[6pt]
&\quad \text{(4) Delta–Gamma–Theta approximation:}\\
&\quad\quad V(S_{t+h}) \approx V(S_t) \;+\; \Delta \,\epsilon \;+\; \tfrac{1}{2}\,\Gamma\,\epsilon^2 \;+\; h\,\theta.
\end{aligned}
\]

## Example: Using the Delta–Gamma Approximation

Suppose at time $t$, the stock price is $S_t = 50$ and the option price is $V(S_t) = 2.50.$  
Assume we know (from the Greeks) that  


$
\Delta = 0.50, \quad \Gamma = 0.02.$

If, over a short interval, the stock price increases by $epsilon = +2$ (i.e., to $52$), we can approximate the new option price $V(S_{t+h})$ using the **Delta–Gamma** approximation:

$
V(S_{t+h})
\,\approx\,
V(S_t)
\;+\;
\Delta\,\epsilon
\;+\;
\tfrac{1}{2}\,\Gamma\,\epsilon^2.
$

Plugging in the numbers:

1. Delta term:  
   $ \Delta \,\epsilon = 0.50 \times 2 = 1.00.$

2. Gamma term:  
   $ \frac{1}{2}\,\Gamma \epsilon^2 = \tfrac{1}{2} (0.02) (2^2) = 0.5 \times 0.02 \times 4 = 0.04. $

3. Total approximate price:  
   $
   V(S_{t+h}) 
   \approx
   2.50 + 1.00 + 0.04 
   \;=\;
   3.54.
   $

So now we have Delta–Gamma approximation, the option price increases from \(2.50\) to approximately \(3.54\) when the stock price moves from \(50\) to \(52\).
