# P2. A
So $2X+Y$ is a linear combination of two normal random variables. Meaning the linear combination's distrobution in a univariate distrobution.

Note
$Var(aX + bY) = a^2Var(X) + b^2Var(Y) + 2abCov(X, Y)$

Thus if $Z = 2X + Y$

Then 

$\mu _Z = 2$

$\sigma _Z ^2 = 4 + 0 + 4Cov(X,Y) = ρ * \sigma _X * \sigma _Y = 4+ 4* (2 * 0.5 * \sqrt {1} * \sqrt {4}) = 12 $


$P(2X + Y \geq 3) = P(Z \leq \frac {(3 - 2)} { (2 \sqrt{3})})$


$P(Z \leq \frac {1} { (2 \sqrt{3})})$


$P(Z \leq 0.2887) $


```python
from scipy.stats import norm
z_score = (12**0.5)**-1
probability = norm.cdf(z_score)
print("Answer:", round(probability, 4))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the three normal distributions
mean1, std1 = 1, 1
mean2, std2 = 0, 4
mean3, std3 = (mean1 *2) + mean2, 12**0.5

# Generate 10,000 data points from each distribution
data1 = np.random.normal(mean1, std1, 100000)
data2 = np.random.normal(mean2, std2, 100000)
data3 = np.random.normal(mean3, std3, 100000)

# Create a histogram for each set of data points
plt.hist(data1, bins=50, density=True, alpha=0.6, color='r', label=f'Mean {mean1}, Std {std1}')
plt.hist(data2, bins=50, density=True, alpha=0.6, color='g', label=f'Mean {mean2}, Std {std2}')
plt.hist(data3, bins=50, density=True, alpha=0.6, color='b', label=f'Mean {mean3}, Std {std3}')

# Add a legend
plt.legend(loc='upper right')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Normal Distribution PDFs')

# Show the plot
plt.show()


print("If we random sample from a normal P(2X+ Y< 3) is: ", sum([(1 if x <=3 else 0)  for x in data3]) /len(data3))  
```
