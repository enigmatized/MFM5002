The heart of problem 1 is you integrate out the other variable:

$f_Y(y)  =  \int _0 ^1 f_{X,Y}(x,y) dx$ 



- This is where a lot of the “heavy lifting” occurs.
- Once you have the marginal PDFs, you can integrate each one from 0 up to x (or y, respectively). This step reconfirms how PDFs and CDFs are related.
- These typically follow from the marginal PDFs if you only want E[X] or E[Y]. But for E[XY], you go back to the joint PDF:

$E[XY]  = \int \int xy f_{X,Y}(x,y) dx dy$
