```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def law_of_total_variance_demo(num_samples=50000, num_groups=5, plot_intermediate_dists=True):
    """
    Demonstrates and visualizes the Law of Total Variance.

    Args:
        num_samples: Number of samples to generate.
        num_groups: Number of discrete groups for the conditioning variable (Y).
        plot_intermediate_dists: Whether to plot distributions of X | Y_i

    Illustrates:
        Var(X) = E[Var(X|Y)] + Var[E(X|Y)]
    """

    # 1.  Define parameters and generate data

    # ---  Parameters for the joint distribution of (X, Y) ---
    # We'll model X | Y = y_i  as  N(mu_i, sigma_i^2)
    # Y will be discrete, taking values from {0, 1, ..., num_groups - 1} with equal probability.

    np.random.seed(42)  # for reproducibility

    group_means = np.random.uniform(2, 18, num_groups) # Randomly chosen means for each group.  These are the E[X|Y=y_i]
    group_std_devs = np.random.uniform(0.5, 3, num_groups)  # Randomly chosen standard deviations.
    group_probs = np.full(num_groups, 1 / num_groups)   # Uniform probabilities for each group (Y values).


    # --- Generate the samples ---
    y_values = np.random.choice(np.arange(num_groups), size=num_samples, p=group_probs)  # Sample Y values.
    x_values = np.zeros(num_samples)

    for i in range(num_groups):
        indices = (y_values == i)  # Find all samples belonging to group i
        num_in_group = np.sum(indices)
        x_values[indices] = np.random.normal(group_means[i], group_std_devs[i], num_in_group) # Generate X values for that group


    # 2. Calculate variances and expectations

    # --- Direct Calculation of Var(X) ---
    overall_variance = np.var(x_values)

    # --- Law of Total Variance ---
    conditional_variances = np.zeros(num_groups)  # E[Var(X|Y)]
    conditional_means = np.zeros(num_groups)     # Var[E(X|Y)]

    for i in range(num_groups):
        indices = (y_values == i)
        conditional_variances[i] = np.var(x_values[indices])  # Var(X | Y = i)
        conditional_means[i] = np.mean(x_values[indices])      # E(X | Y = i)

    expected_conditional_variance = np.mean(conditional_variances)  # E[Var(X|Y)]
    variance_of_conditional_means = np.var(conditional_means)      # Var[E(X|Y)]

    lotv_variance = expected_conditional_variance + variance_of_conditional_means

    print(f"Directly Calculated Variance of X: {overall_variance:.4f}")
    print(f"Variance from Law of Total Variance: {lotv_variance:.4f}")
    print(f"   E[Var(X|Y)]: {expected_conditional_variance:.4f}")
    print(f"   Var[E(X|Y)]: {variance_of_conditional_means:.4f}")



    # 3.  Create visualizations
    if not plot_intermediate_dists:
      return

    fig, axes = plt.subplots(2, 1, figsize=(12, 10), gridspec_kw={'height_ratios': [3, 2]})  # Two subplots, one for the distributions, one for variance breakdown
    ax_dist = axes[0] # top
    ax_var = axes[1] # bottom

    # --- Plot overall distribution of X ---
    x_range = np.linspace(x_values.min() - 1, x_values.max() + 1, 500)
    overall_pdf = stats.norm.pdf(x_range, np.mean(x_values), np.std(x_values))
    ax_dist.plot(x_range, overall_pdf, color='black', linewidth=2, label='Overall X Distribution')



    # --- Plot conditional distributions X | Y_i, and show means and variances ---
    colors = plt.cm.viridis(np.linspace(0, 1, num_groups))  # Nice colormap

    for i in range(num_groups):
        indices = (y_values == i)
        x_given_y = x_values[indices]
        conditional_pdf = stats.norm.pdf(x_range, group_means[i], group_std_devs[i])

        # Conditional distribution, scaled and shifted for visualization
        ax_dist.plot(x_range, conditional_pdf * 0.2 + i, color=colors[i], linewidth=1.5)

        #  E[X|Y]
        ax_dist.vlines(conditional_means[i], i, i + 0.2, color=colors[i], linestyles='dashed', linewidth=2)
        ax_dist.text(conditional_means[i] + 0.2, i + 0.15, f"E[X|Y={i}]", color=colors[i], fontsize=9)

        #  Var(X|Y) -  visualized as horizontal bars
        var_start = conditional_means[i] - np.sqrt(conditional_variances[i])
        var_end = conditional_means[i] + np.sqrt(conditional_variances[i])
        ax_dist.hlines(i + 0.05, var_start, var_end, color=colors[i], linewidth=3)
        ax_dist.text(var_end + 0.2, i + 0.05, f"V[X|Y={i}]", color=colors[i], fontsize=9)

    ax_dist.set_xlabel("X")
    ax_dist.set_ylabel("Y (Group)")
    ax_dist.set_yticks(np.arange(num_groups))
    ax_dist.set_yticklabels([f"Y={i}" for i in range(num_groups)])
    ax_dist.set_title("Conditional Distributions and Overall Distribution of X")
    ax_dist.legend()


    # --- Bar plot for the variance breakdown ---
    bar_width = 0.35
    ax_var.bar(0, expected_conditional_variance, bar_width, label='E[Var(X|Y)]', color='skyblue')
    ax_var.bar(bar_width, variance_of_conditional_means, bar_width, label='Var[E(X|Y)]', color='coral')

    ax_var.bar(0 + 2*bar_width, overall_variance, bar_width, label='Var(X) [Direct]', color='gray', alpha=0.5)
    ax_var.bar(0 + 3*bar_width, lotv_variance, bar_width, label='Var(X) [LOTV]', color='green', alpha=0.5)

    ax_var.set_xticks([bar_width/2, 3*bar_width/2, 5*bar_width/2, 7*bar_width/2])
    ax_var.set_xticklabels(['E[Var(X|Y)]', 'Var[E(X|Y)]', 'Var(X) [Direct]', 'Var(X) [LOTV]'])
    ax_var.set_ylabel("Variance")
    ax_var.set_title("Law of Total Variance Breakdown")
    ax_var.legend()

    plt.tight_layout()
    plt.show()


law_of_total_variance_demo()

law_of_total_variance_demo(num_groups=8, plot_intermediate_dists=False)  # Another example, skipping the distributions
```
