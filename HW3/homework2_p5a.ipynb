$f(x,y) = \begin{cases}
 \frac{3}{2}y^2 & 0 \le x \le 2 \text{ and } 0 \le y \le 1 \ 
  &  \\ 0 & \text{otherwise} \end{cases} $




### Determine the joint CDF of X and Y.

So what makes this tricky is that there are mulitple bounds to consider.
The easy one is

$F_{X,Y}​(x,y)=\int _0 ^x ​\int _0 ^y \frac{3}{2}y^2 dx dy  =\frac{xy^3}{2}​ $

But that only applies to the bounds of (0,2) for x and (0, 1) for y.

##### We have 6 cases to consider

1. If $x<0$ 
2.  $y<0$
3. $ 0 \leq x \leq 2 $ and $ 0 \leq y \leq 1$ 
4. $x \geq 2$  and  $ 0 \leq y \leq 1$ 
5.   $ 0 \leq x \leq 2 $ and $y \geq 1$
6. $y \geq 1$ and $x \geq 2$

#### Case 1+2
$F_{X,Y}​(x,y)=0$

#### Case 3
   $$
   \begin{aligned}
   F_{X,Y}(x,y) 
   &= 
   \int_{u=0}^{x} \int_{v=0}^{y} \tfrac{3}{2}\,y^{2}\,\mathrm{d}y\,\mathrm{d}x 
   \\[6pt]
   &= 
   \int_{x=0}^{x}
     \Bigl[\tfrac{3}{2}\,\tfrac{y^{3}}{3}\Bigr]_{y=0}^{y=y}
   \,\mathrm{d}x
   \;=\;
   \int_{x=0}^{x} 
     \tfrac{y^{3}}{2}
   \,\mathrm{d}x
   \;=\;
   \tfrac{x\,y^{3}}{2}.
   \end{aligned}
   $$


#### Case 4 - $x > 2$ but $0 \le y \le 1$
   $$
   F_{X,Y}(x,y)
   \;=\;
   \int_{x=0}^{2} \int_{y=0}^{y} \tfrac{3}{2}\,y^{2}\,\mathrm{d}y\,\mathrm{d}x
   =
   2 \,\cdot
   \Bigl[\tfrac{3}{2}\,\tfrac{y^{3}}{3}\Bigr]_{0}^{y}
   =
   2 \,\cdot \tfrac{y^{3}}{2}
   =
   y^{3}.
   $$

#### Case 5 - $y > 1$ but $0 \le x \le 2$
   $$
   F_{X,Y}(x,y)
   \;=\;
   \int_{x=0}^{x} \int_{y=0}^{1} \tfrac{3}{2}\,y^{2}\,\mathrm{d}y\,\mathrm{d}x
   =
   \int_{x=0}^{x}
     \Bigl[\tfrac{3}{2}\,\tfrac{y^{3}}{3}\Bigr]_{0}^{1}
   \,\mathrm{d}x
   =
   \int_{x=0}^{x} \tfrac{1}{2}\,\mathrm{d}x
   =
   \tfrac{x}{2}.
   $$

#### Case 6  - $x > 2$ and $y > 1$
   $$
   F_{X,Y}(x,y) = 1
   $$



Putting it all together, the **joint CDF** is:

$$
F_{X,Y}(x,y)
=
\begin{cases}
0, 
& x < 0 \;\text{or}\; y < 0,\\[6pt]
\dfrac{x\,y^{3}}{2}, 
& 0 \le x \le 2, \; 0 \le y \le 1,\\[6pt]
y^{3},
& x > 2, \; 0 \le y \le 1,\\[6pt]
\dfrac{x}{2},
& 0 \le x \le 2, \; y > 1,\\[6pt]
1,
& x > 2, \; y > 1.
\end{cases}
$$

### **(b) Marginal PDFs $f_X(x)$ and $f_Y(y)$**

#### Marginal of $X$

$$
\begin{aligned}
f_X(x)
&=
\int_{y=0}^{1} f_{X,Y}(x,y)\,\mathrm{d}y
=
\int_{0}^{1} \tfrac{3}{2}\,y^{2}\,\mathrm{d}y
\\[6pt]
&=
\tfrac{3}{2}
\left[\dfrac{y^{3}}{3}\right]_{0}^{1}
=
\tfrac{3}{2}\,\cdot \dfrac{1}{3}
=
\dfrac{1}{2},
\end{aligned}
$$

valid for $0 \le x \le 2$, and $0$ otherwise. Hence:

$$
f_X(x) 
=
\begin{cases}
\dfrac12, & 0 \le x \le 2,\\
0, & \text{otherwise}.
\end{cases}
$$

#### Marginal of \(Y\)

$$
\begin{aligned}
f_Y(y)
&=
\int_{x=0}^{2} f_{X,Y}(x,y)\,\mathrm{d}x
=
\int_{0}^{2} \tfrac{3}{2}\,y^{2}\,\mathrm{d}x
\\[6pt]
&=
\tfrac{3}{2}\,y^{2} 
\int_{0}^{2}\mathrm{d}x
=
\tfrac{3}{2}\,y^{2} \cdot 2
=
3\,y^{2},
\end{aligned}
$$

valid for $0 \le y \le 1$, and $0$ otherwise. Therefore:

$$
f_Y(y) 
=
\begin{cases}
3\,y^{2}, & 0 \le y \le 1,\\
0, & \text{otherwise}.
\end{cases}
$$

## Part G Conditional PMF of X given X + Y = 3


$
f_{X|X+Y}(x|3) = \begin{cases}
0, & \text{if } x = 0 \\
\frac{1}{3}, & \text{if } x = 1 \\
\frac{1}{3}, & \text{if } x = 2 \\
\frac{3}{7}, & \text{if } x = 3 \\
0, & \text{otherwise}
\end{cases}
$


We want $f_{X|X+Y} P(x|3) = P(X=x | X+Y=3) = P(X=x, X+Y=3) / P(X+Y=3)$. 

$P(X+Y=3) = 0.42$.  


$f_{X|X+Y}(0|3) = P(0,3)/0.42 = 0$


$f_{X|X+Y}(1|3) = P(1, 2) / 0.42 = 0.14 / 0.42 = 1/3$

$f_{X|X+Y}(2|3) = P(2, 1) / 0.42 = 0.14 / 0.42 = 1/3$

$f_{X|X+Y}(3|3) = P(3, 0) / 0.42 = 0.18 / 0.42 = 3/7$


This is equivalent to $P(X=x, Y=3-x) / P(X+Y=3)$.
