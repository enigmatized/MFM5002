# GAME 1


Let $X$ be the value of a fair six-sided die roll, so $X \in \{1,2,3,4,5,6\}$ with probability $\frac{1}{6}$ each.

Let $Y$ be the number of heads in 4 fair coin tosses, so $Y \sim \text{Binomial}(n=4, p=\frac{1}{2})$. That is, $Y \in \{0,1,2,3,4\}$ with
$
\mathbb{P}(Y = k) = \binom{4}{k} \left(\frac{1}{2}\right)^4 = \frac{\binom{4}{k}}{16}.
$

Define $Z = X^Y$. We want $\mathbb{E}[Z]$.

1. **Use the Law of Total Expectation**  
   Since $X$ and $Y$ are independent, we get:
   $
   \mathbb{E}[Z] 
   = \mathbb{E}[X^Y]
   = \sum_{k=0}^4 \mathbb{E}[X^k] \, \mathbb{P}(Y = k).
   $


   Therefore:
   
   
   $
   \mathbb{E}[Z] = \sum_{k=0}^4 \binom{4}{k} \,\frac{1}{16} \,\mathbb{E}[X^k].
   $

2. **Compute $\mathbb{E}[X^k]$ for $k = 0, 1, 2, 3, 4$**  
   For a fair die $X$ uniform on $\{1, 2, 3, 4, 5, 6\}$:
   - $\mathbb{E}[X^0] = 1$.
   - $\mathbb{E}[X^1] = \frac{1+2+3+4+5+6}{6} = 3.5$.
   - $\mathbb{E}[X^2] = \frac{1^2 + 2^2 + 3^2 + 4^2 + 5^2 + 6^2}{6} = \frac{91}{6} \approx 15.1667.$
   - $\mathbb{E}[X^3] = \frac{1^3 + 2^3 + 3^3 + 4^3 + 5^3 + 6^3}{6} = \frac{441}{6} = 73.5.$
   - $\mathbb{E}[X^4] = \frac{1^4 + 2^4 + 3^4 + 4^4 + 5^4 + 6^4}{6} = \frac{2275}{6} \approx 379.1667.$

3. **Combine with Binomial coefficients**  
   Since $\binom{4}{0} = 1$, $\binom{4}{1} = 4$, $\binom{4}{2} = 6$, $\binom{4}{3} = 4$, and $\binom{4}{4} = 1$
   :

   $   \mathbb{E}[Z] = \frac{1}{16} \bigl[ \binom{4}{0}\,\mathbb{E}[X^0] + \binom{4}{1}\,\mathbb{E}[X^1] + \binom{4}{2}\,\mathbb{E}[X^2] + \binom{4}{3}\,\mathbb{E}[X^3]      + \binom{4}{4}\,\mathbb{E}[X^4] \bigr].
   $

   
   Substituting values:
   $ \mathbb{E}[Z] = \frac{1}{16}\Bigl[ 1 \cdot 1 + 4 \cdot 3.5  + 6 \cdot 15.1667 + 4 \cdot 73.5 + 1 \cdot 379.1667 \Bigr] $.
   
   
   Numerically, that sum is approximately $48.6979$.

4. **Final Answer**  
   $
   \boxed{\mathbb{E}[Z] \approx 48.70.}
   $


## Law of Total Variance

Use:
$
\mathbb{Var}(Z) = \mathbb{E}\bigl[\mathrm{Var}(Z \mid Y)\bigr] + \mathrm{Var}\bigl(\mathbb{E}[Z \mid Y]\bigr).
$

Since $X$ and $Y$ are independent:

1. $\mathbb{E}[Z \mid Y=k] = \mathbb{E}[X^k]$
2. $\mathrm{Var}(Z \mid Y=k) = \mathrm{Var}(X^k) = \mathbb{E}[X^{2k}] - \bigl(\mathbb{E}[X^k]\bigr)^2.$

Hence:

$ \mathrm{Var}(Z) = \underbrace{\mathbb{E}\bigl[\mathrm{Var}(Z \mid Y)\bigr]}_{\text{Term 1}} + \underbrace{\mathrm{Var}\bigl(\mathbb{E}[Z \mid Y]\bigr)} {\text{Term 2}}.
$

---

### Term 1: $\mathbb{E}[\mathrm{Var}(Z \mid Y)]$

$ \mathbb{E} \bigl[\mathrm{Var}(Z \mid Y)\bigr] = \sum_{k=0}^4 \Bigl[\mathbb{E}[X^{2k}] - \bigl(\mathbb{E}[X^k]\bigr)^2 \Bigr] \;\mathbb{P}(Y = k).
$

### Term 2: $\mathrm{Var}\bigl(\mathbb{E}[Z \mid Y]\bigr)$

$ \mathrm{Var}\bigl(\mathbb{E}[Z \mid Y]\bigr) = \mathrm{Var}\bigl(\mathbb{E}[X^Y]\bigr) = \mathrm{Var}\bigl(M_Y \bigr), $

where $M_k = \mathbb{E}[X^k]$. Then

$
\mathrm{Var}(M_Y) = \sum_{k=0}^4 \bigl(M_k\bigr)^2 \,\mathbb{P}(Y=k) \;-\; \Bigl[\sum_{k=0}^4 M_k \,\mathbb{P}(Y=k)\Bigr]^2.
$

Putting them together:

$
\mathrm{Var}(Z) = \sum_{k=0}^4 \Bigl[\mathbb{E}[X^{2k}] - \bigl(\mathbb{E}[X^k]\bigr)^2 \Bigr] \,\mathbb{P}(Y=k) \;+\; \sum_{k=0}^4 \bigl(M_k\bigr)^2 \,\mathbb{P}(Y=k) \;-\; \Bigl[\sum_{k=0}^4 M_k \,\mathbb{P}(Y=k)\Bigr]^2. $

---

## Computing $\mathbb{E}[X^k]$ for a Fair Die

Define
$
M_k = \mathbb{E}[X^k] = \frac{1^k + 2^k + 3^k + 4^k + 5^k + 6^k}{6}.
$
For example:
- $M_0 = 1.$
- $M_1 = 3.5.$
- $M_2 = \frac{91}{6} \approx 15.1667.$
- $M_3 = \frac{441}{6} = 73.5.$
- $M_4 = \frac{2275}{6} \approx 379.1667.$
- $M_5, M_6, \dots$ similarly, as needed up to $M_8$.

---




```python
# We'll compute E[X^k] for k = 0..8 for a fair die X
import math

import itertools

# All powers we might need
def E_Xk(k):
    return sum(n**k for n in range(1,7))/6

M = [E_Xk(k) for k in range(9)]  # 0..8

# Probability that Y=k for k=0..4
from math import comb
p = [comb(4,k)/16 for k in range(5)]

# 1) E[Var(Z|Y)] = sum_{k=0..4} [ M_{2k} - (M_k)^2 ] * p[k]
term1 = sum( (M[2*k] - M[k]**2)*p[k] for k in range(5) )

# 2) Var(E[Z|Y]) = sum_{k=0..4} (M_k)^2 * p[k] - [sum_{k=0..4} M_k * p[k]]^2
m1 = sum(M[k]*p[k] for k in range(5))           # E[M_Y]
m2 = sum((M[k]**2)*p[k] for k in range(5))      # E[M_Y^2]
term2 = m2 - m1**2

varZ = term1 + term2

print("Var(Z) =", varZ)
```
