
#### Two-Stage Game Setup

We have a game $G$ defined as follows:

1. Flip a fair coin (probability $\tfrac12$ each of Heads/Tails).
   - If the first flip is Heads, flip again:
     - If the second flip is also Heads: you lose 100 immediately, i.e. $ \textbf{payout} = -100$.
     - Otherwise (Heads then Tails): proceed to Game 1.
   - If the first flip is Tails: proceed to Game 1 immediately.

2. $\textbf{Game 1}$:
   - Roll a fair six-sided die, call the result $X \in \{1,2,3,4,5,6\}$.
   - Flip a fair coin 4 times and let $Y$ be the number of heads (so $Y \sim \mathrm{Binomial}(4,\tfrac12)$).
   - The payout from Game 1 is $Z = X^Y$.



Overall, the payout $G$ is:
$
G \;=\;  \begin{cases} -100, & \text{with probability } \tfrac{1}{4},\
Z, & \text{with probability } \tfrac{3}{4},
\end{cases}
$

because two consecutive heads (probability $\tfrac12 \times \tfrac12 = \tfrac14$) leads to an immediate $-100$ payout, and otherwise ($\tfrac34$ chance) you get $Z$.





#### Expected Value of the Two-Stage Game.


[To calculate of Game1 or $X^Y$](https://github.com/enigmatized/MFM5002/blob/main/Project1/Game1.ipynb)

We already know the expectation of $Z$, $\mathbb{E}[Z]\approx 48.70$. Therefore:
$
\mathbb{E}[G]
\;=\;
\underbrace{\bigl(\tfrac14\bigr)\times(-100)}_{\text{lose 100}}
\;+\;
\underbrace{\bigl(\tfrac34\bigr)\times \mathbb{E}[Z]}_{\text{play Game 1}}
\;=\;
-25 
\;+\;
\tfrac34 \times 48.70
\;=\;
-25 
\;+\;
36.525
\;=\;
11.525.
$




Simulations give about $11.47$ or $11.48$, which is very close.

#### Variance of the Two-Stage Game (Direct Second-Moment Approach)

NOTE:
$
\mathrm{Var}(G) 
\;=\; 
\mathbb{E}[G^2] 
\;-\; 
\bigl(\mathbb{E}[G]\bigr)^2.
$


Since $G$ is either $-100$ (with prob.$\,\tfrac14$) or $Z$ (with prob.$\,\tfrac34$), we have


$
G^2 
\;=\;
\begin{cases}
10000, & \text{with prob. } \tfrac14,\\
Z^2,   & \text{with prob. } \tfrac34.
\end{cases}
$


Hence,


$
\mathbb{E}[G^2] 
\;=\; 
\bigl(\tfrac14\bigr)\times 10000 
\;+\; 
\bigl(\tfrac34\bigr)\times \mathbb{E}[Z^2].
$


But $\mathbb{E}[Z^2] = \mathrm{Var}(Z) + (\mathbb{E}[Z])^2$. 


From Game 1, $\mathrm{Var}(Z)\approx 22892$ and $\mathbb{E}[Z]\approx 48.70$. 


Thus
$
\mathbb{E}[Z^2] 
\;=\; 
22892 + (48.70)^2 
\;=\; 
22892 + 2376.49
\;=\; 
25268.49.
$


So
$
\mathbb{E}[G^2]
\;=\;
\tfrac14\times 10000
\;+\;
\tfrac34\times 25268.49
\;=\;
2500
\;+\;
18951.3675
\;=\;
21451.3675.
$



We already had $\mathbb{E}[G]\approx 11.525$. 


Therefore,
$
\mathrm{Var}(G)
\;=\;
21451.3675
\;-\;
(11.525)^2
\;=\;
21451.3675
\;-\;
132.90625
\;=\;
21318.46125.
$



In simulation I found around $21228.8$ or so. 

Quite close to this theoretical value.



#### Variance of the Two-Stage Game (Law of Total Variance)

Let $T$ be the event/indicator that the game proceeds to Game 1 (i.e.\ not two heads in a row). 

So:
$ T  \;=\; \begin{cases} 0, & \text{if we get two heads and lose 100,}\\ 1, & \text{otherwise (play Game 1).} \end{cases} $

Then $\Pr(T=0)=\tfrac14$, $\Pr(T=1)=\tfrac34$, and

$ G  \;=\; -100 \cdot \mathbf{1}_{\{T=0\}}  \;+\; Z \cdot \mathbf{1}_{\{T=1\}}. $

Use the law of total variance conditioning on $T$:

$ \mathrm{Var}(G) \;=\; \underbrace{\mathbb{E}[\mathrm{Var}(G \mid T)]}_{\text{Term 1}} \;+\; \underbrace{\mathrm{Var}(\mathbb{E}[G \mid T])}_{\text{Term 2}}.
$



##### Term 1: $\mathbb{E}[\mathrm{Var}(G\mid T)]$.

- If $T=0$, then $G=-100$ with probability 1, so $\mathrm{Var}(G\mid T=0)=0$.
- If $T=1$, then $G=Z$, so $\mathrm{Var}(G\mid T=1)=\mathrm{Var}(Z)$.

Hence:
$
\mathbb{E}[\mathrm{Var}(G \mid T)]
\;=\;
(\tfrac14)\times 0 
\;+\;
(\tfrac34)\times \mathrm{Var}(Z)
\;=\;
\tfrac34\,\mathrm{Var}(Z).
$



#### Term 2: $\mathrm{Var}(\mathbb{E}[G\mid T])$.

- $\mathbb{E}[G\mid T=0] = -100.$
- $\mathbb{E}[G\mid T=1] = \mathbb{E}[Z] \approx 48.70.$

So $\mathbb{E}[G\mid T]$ is a random variable taking value $-100$ with probability $\tfrac14$ and $48.70$ with probability $\tfrac34$.

Its variance is

$ \mathrm{Var}\bigl(\mathbb{E}[G\mid T]\bigr) \;=\; (\tfrac14)\times(-100)^2 \;+\; (\tfrac34)\times(48.70)^2
\;-\;
\Bigl[(\tfrac14)\times(-100)+( \tfrac34)\times(48.70)\Bigr]^2.
$


Numerically, this yields a few thousand, matching the difference between $\tfrac34\,\mathrm{Var}(Z)$ and the total variance found earlier.


The final result via the law of total variance is the same $\mathrm{Var}(G)\approx 21300+$ range before.
