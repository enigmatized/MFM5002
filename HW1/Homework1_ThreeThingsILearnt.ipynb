```python
#Chat gpt code helping me out

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 1) Define a fixed mean:
mu = 10.0

# 2) Define some CV values:
cv_list = [0.1, 0.5, 1.0, 2.0]

# 3) Compute corresponding standard deviations:
sigma_list = [cv * mu for cv in cv_list]

# --- PLOT 1: Bar chart of CV vs. sigma ---

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

ax1 = axes[0]
x_positions = np.arange(len(cv_list))

ax1.bar(x_positions, sigma_list, width=0.6, alpha=0.7, color='skyblue')
ax1.set_xticks(x_positions)
ax1.set_xticklabels([f"CV={cv:.1f}" for cv in cv_list])
ax1.set_ylabel("Standard Deviation ()")
ax1.set_title("Different  Values for Each CV (=10)")

# Annotate the bars with the numeric sigma values:
for i, (cv, sig) in enumerate(zip(cv_list, sigma_list)):
    ax1.text(i, sig + 0.1, f" = {sig:.1f}", ha='center', va='bottom')

# --- PLOT 2: PDF curves for each Normal distribution ---

ax2 = axes[1]

# Define an x-range wide enough to see the distributions
x = np.linspace(0, 50, 400)

for cv, sig in zip(cv_list, sigma_list):
    # Create the pdf for Normal(mean=mu, std=sig)
    pdf = norm.pdf(x, loc=mu, scale=sig)
    label_str = f"CV={cv:.1f}, ={sig:.1f}"
    ax2.plot(x, pdf, label=label_str, linewidth=2)

ax2.set_xlabel("x")
ax2.set_ylabel("PDF")
ax2.set_title("Normal Distributions with Different CV")
ax2.legend()

plt.tight_layout()
plt.show()
```

# Stationary Coefficient of Variation in Lognormal Distributions

## Lognormal Setup: Let $X = e^Y$, where $Y \sim \mathcal{N}(\mu, \sigma^2)$.
Then $X$ is lognormally distributed.

## Mean and Variance

$
\mathbb{E}[X] = e^{\mu + \tfrac{\sigma^2}{2}}, 
\quad
\mathrm{Var}(X) = e^{2\mu + \sigma^2}\,\bigl(e^{\sigma^2} - 1\bigr).
$
Thus
$
\sqrt{\mathrm{Var}(X)} 
= e^{\mu + \tfrac{\sigma^2}{2}}\,\sqrt{e^{\sigma^2} - 1}.
$

\subsection*{Coefficient of Variation (CV)}

$
\text{CV}(X) 
= \frac{\sqrt{\mathrm{Var}(X)}}{\mathbb{E}[X]}
= \frac{e^{\mu + \tfrac{\sigma^2}{2}} \sqrt{e^{\sigma^2} - 1}}{e^{\mu + \tfrac{\sigma^2}{2}}}
= \sqrt{e^{\sigma^2} - 1}.
$

Notice that $\text{CV}(X)$ only depends on $\sigma$, not $\mu$. This means if we shift $\mu$, 
the distribution's mean changes but its CV remains constant.

In contrast, SD varies depending upon the expected value of measurements.

A more robust possibility is the quartile coefficient of dispersion, half the interquartile range ( Q 3  Q 1 ) / 2.
