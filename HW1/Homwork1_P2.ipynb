```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Generating a normal distribution
data_normal = norm.rvs(loc=0, scale=1, size=1000)

# Calculating statistics
mean_normal = np.mean(data_normal)
median_normal = np.median(data_normal)
std_normal = np.std(data_normal)
iqr_normal = np.percentile(data_normal, 75) - np.percentile(data_normal, 25)

print("Normal Distribution")
print(f"Mean: {mean_normal}, Median: {median_normal}, STD: {std_normal}, IQR: {iqr_normal}")

# Plotting
plt.hist(data_normal, bins=30, alpha=0.7, color='blue')
plt.title('Normal Distribution')
plt.show()
```

IQR (Interquartile Range) vs. Standard Deviation (SD)

1. Definition
   - **IQR**: The difference between the 75th percentile (Q3) and the 25th percentile (Q1) of the data.
   - **SD**: A measure of how spread out the data is from the mean, calculated using the squared deviations from the mean.

2. Robustness to Outliers
   - **IQR**: Robust against outliers. Because it relies on median-based statistics (Q1 and Q3), extreme values have minimal influence.
   - **SD**: Sensitive to outliers. Extreme values can greatly inflate the SD because of the squared term in its formula.

3. Data Distribution
   - **IQR**: More informative when data is skewed or has heavy tails, as it provides a central range of the data that isn’t unduly affected by outliers.
   - **SD**: Most informative for approximately symmetric (e.g., normal) distributions where the mean is a good central tendency measure.

4. When to Use
   - **Use IQR**:
     - When the dataset is not normally distributed or is heavily skewed.
     - When outliers are present and you need a robust measure of spread.
   - **Use SD**:
     - When the data is roughly normal or symmetric.
     - When you’re performing analyses (like parametric tests) that assume normality.
     - When the mean is a good measure of central tendency.


```python
import numpy as np
from scipy.stats import iqr

# Example dataset
data = [10, 12, 15, 18, 20, 21, 21, 22, 50]  # Notice the outlier at 50

# Calculate IQR
data_iqr = iqr(data)
print("IQR:", data_iqr)

# Calculate Standard Deviation
data_sd = np.std(data, ddof=1)  # Using sample standard deviation with ddof=1
print("SD:", data_sd)
```

Median VS Mean

Use Median when:

- The distribution is skewed (asymmetric)
- There are significant outliers
- The data has long tails

Real-world examples for median:


- Income distributions (many people make moderate salaries, few make extremely high amounts)
- Response times for web servers (occasional very long responses skew the distribution)

Use Mean when:

- The distribution is symmetric
- Outliers are rare or non-existent
- You want to account for all values equally




The key principle is: Mean is heavily influenced by extreme values, while median is resistant to them. 
That's why it is said "median is robust to outliers."
For most physical measurements that follow a normal (Gaussian) distribution, mean is the go to.
