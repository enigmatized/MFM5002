## The Probability Integral Transform

### Statement -  Let $X$ be a random variable with CDF $F$.

Define 
$
Y = F(X).
$
Then $Y$ is Uniform(0,1) distributed.

#### Proof 

We want to show $P(Y \le y) = y$ for $0 \le y \le 1$.

$
P(Y \le y) 
= P\bigl(F(X) \le y\bigr).
$
Consider three cases:

1. If $y \le 0$, then $P(F(X) \le y) = 0$.

2. If $y \ge 1$, then $P(F(X) \le y) = 1$.

3. If $0 < y < 1$, let $F^{-1}(y) = \inf\{x: F(x) \ge y\}$. Then
$
P(F(X) \le y) 
= P\bigl(X \le F^{-1}(y)\bigr)
= F\bigl(F^{-1}(y)\bigr).
$


 - If $F$ is continuous and strictly increasing, this equals $y$. In general, by properties of distribution functions and quantile functions, this is still $\le y$ and the detailed definition shows it coincides with $y$ in the continuous case.

Hence, the CDF of $Y$ is
$
P(Y \le y) =
\begin{cases}
0, & y \le 0,\\
y, & 0 \le y \le 1,\\
1, & y \ge 1.
\end{cases}
$
This is precisely the CDF of a $\text{Uniform}(0,1)$ distribution.

Intuition: Why Is It Uniform?

    The transformation y=F(x) “pushes” all of the random variable’s possible values (according to its distribution F) into the range [0,1].
    Because F is a CDF, it is nondecreasing and takes values between 0 and 1. Thus Y=F(X) is automatically confined to [0,1].
    
    
    The fundamental reason it becomes uniform is that each quantile region of X is “stretched” or “compressed” so that they all have the same probability mass once transformed by F.

This result is extremely useful in generating random variables with a specified distribution: if you can generate a Uniform(0,1) random variable U, then $X=F^{−1}(U)$ has the CDF F.
