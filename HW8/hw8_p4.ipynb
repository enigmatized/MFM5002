```python
vecctors = []
for val in eigenvalues:
    print(f"\nEigenvalue λ = {val}:")
    eig_matrix = A - val * sp.eye(3)
    null_space = eig_matrix.nullspace()

    if null_space:
        for i, vec in enumerate(null_space):
            print(f"Eigenvector {i + 1}:")
            sp.pprint(vec)
            vecctors.append(vec)
    else:
        print("No eigenvector found (should not happen).")
```

```python
def gram_schmidt(vectors):
    orthonormal_basis = []
    
    for v in vectors:
        w = v.copy()
        for u in orthonormal_basis:
            proj = (v.dot(u)) * u
            w -= proj
        w = sp.simplify(w / w.norm())
        orthonormal_basis.append(w)

    return orthonormal_basis


orthonormal = gram_schmidt(vecctors)

for i, vec in enumerate(orthonormal, 1):
    print(f"u{i} = {vec}")
```

```python
import sympy as sp

# Define the matrix
A = sp.Matrix([
    [3, -2, 4],
    [-2, 6, 2],
    [4, 2, 4]
])

# Step 1: Find eigenvalues and eigenvectors
eigen_data = A.eigenvects()

# Step 2: Flatten all eigenvectors and apply Gram-Schmidt
vecs = []
for eigenval, _, vect_list in eigen_data:
    vecs.extend(vect_list)

# Step 3: Gram-Schmidt Orthonormalization
def gram_schmidt(vectors):
    ortho = []
    for v in vectors:
        for u in ortho:
            v -= v.project(u)
        ortho.append(v)
    return [v / v.norm() for v in ortho]

ortho_vecs = gram_schmidt(vecs)

# Step 4: Construct matrix P from orthonormal eigenvectors
P = sp.Matrix.hstack(*ortho_vecs)

# Step 5: Compute diagonal matrix D
D = P.T * A * P

# Show results
sp.pprint(D)
```

```python
import numpy as np

# Step 1: Define the matrix
A = np.array([
    [3, -2, 4],
    [-2, 6, 2],
    [4, 2, 4]
], dtype=float)

# Step 2: Get eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eigh(A)

# Step 3: Normalize eigenvectors (they're already orthonormal from eigh, but we’ll ensure it)
Q = eigenvectors
Q = Q / np.linalg.norm(Q, axis=0)

# Step 4: Create diagonal matrix D
D = np.dot(np.dot(Q.T, A), Q)

# Print results
print("Orthonormal Matrix P (Q):")
print(np.round(Q, 4))

print("\nDiagonal Matrix D:")
print(np.round(D, 4))
```
