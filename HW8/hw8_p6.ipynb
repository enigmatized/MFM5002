```python
import sympy as sp

λ = sp.Symbol('λ')

A = sp.Matrix([
    [6, -2, -1],
    [-2, 6, -1],
    [-1, -1, 5]
])

char_poly = A.charpoly(λ)


print("Characteristic Polynomial:")
print(char_poly.as_expr())

λ = sp.Symbol('λ')
poly = λ**3 - 13*λ**2 + 30*λ + 84

# Factor the polynomial
factored = sp.factor(char_poly)
print(factored)



# Define the variable λ
λ = sp.symbols('λ')

# Create the characteristic polynomial: det(A - λI)
char_poly = (A - λ * sp.eye(3)).det()

# Expand the polynomial (optional but nice for viewing)
char_poly = sp.expand(char_poly)
print("Characteristic polynomial:")
sp.pprint(char_poly)

# Solve for eigenvalues (roots of the characteristic polynomial)
eigenvalues = sp.solve(char_poly, λ)
print("\nEigenvalues:")
print(eigenvalues)

vecctors = []
for val in eigenvalues:
    print(f"\nEigenvalue λ = {val}:")
    eig_matrix = A - val * sp.eye(3)
    null_space = eig_matrix.nullspace()

    if null_space:
        for i, vec in enumerate(null_space):
            print(f"Eigenvector {i + 1}:")
            sp.pprint(vec)
            vecctors.append(vec)
    else:
        print("No eigenvector found (should not happen).")
```
