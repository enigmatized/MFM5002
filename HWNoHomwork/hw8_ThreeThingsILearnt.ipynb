```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define a 2D matrix A
A = np.array([[3, 1],
              [0, 2]])

# Step 2: Make a grid of vectors to show the space
X, Y = np.meshgrid(np.linspace(-2, 2, 20), np.linspace(-2, 2, 20))
original_vectors = np.stack([X, Y], axis=-1)  # shape (20, 20, 2)

# Step 3: Apply A to every vector in the grid
transformed_vectors = original_vectors @ A.T

# Step 4: Plot the original and transformed vector fields
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot original grid
axes[0].quiver(X, Y, original_vectors[..., 0], original_vectors[..., 1], color='gray')
axes[0].set_title('Original Grid')
axes[0].axis('equal')
axes[0].grid(True)

# Plot transformed grid
axes[1].quiver(transformed_vectors[..., 0], transformed_vectors[..., 1],
               original_vectors[..., 0], original_vectors[..., 1], color='red')
axes[1].set_title('Transformed Grid (A applied)')
axes[1].axis('equal')
axes[1].grid(True)

plt.tight_layout()
plt.show()

# Step 5: Diagonalize A
eigvals, eigvecs = np.linalg.eig(A)
print("Eigenvalues:", eigvals)
print("Eigenvectors (columns):\n", eigvecs)
```
