7. What does it mean to be a span?
A span is simply is the set of linear independent vectors in a sub vector space.
Meaning if we are in the 2-d world, the span of most pair of 2d vectors is all of 2d space.
Thinking about it another way.
x$v_1$ + y$w$ where we can make any combinatins with multiplying vector $v$ with some constant and adding it with vector $w$ times some constant.
In that frame it is easy to imagine "spanning" across a space.
Note when 2-d vectors line up, they are all vectors that line up on a certain line. So they are still in 2d, but only span a line.


Note something that is helpful, is the fact that if we have 2 vectors in a 3-d space, their span will be a plane in the 2-d space.
You can also have three vectors in a 3-d space, but then they span the entire space, again, unless two or more of the three vectors are the same vector.

8. What is a sub space of a matrix?

A subspace of $R^n$ is any set of vectors that is closed under addition and scalar multiplication.

9. What is the basis of a matrix?

A basis is a set of linearly independent vectors that span a vector space.

10. What is the dimension of a matrix?
The number of linearly independent vectors.
Moral of the story here, you cannot have less vectors than dimensions.
Also every basis has the same number, which is the dimension of the sub-space.

11. Connection of independence, basis and dimension.

If you have a bunch of independent vectors, then you are saying you cannot reduce any of the vectors, or two of those vectors are equivilent.
They are all unique and their combinations are unique.

Basis is a subspace of a vector basis. It maybe the subspace of a set of linearly independent vectors.
The number of those vectors is the dimension.

Note, something easy to confuse.
If you have a basis of dimension of 2, it does not mean $R^2$.
It can be a plane in $R^3$ with a dimension of 2.
