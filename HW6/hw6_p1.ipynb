## A

**Given system of equations:**

1.  $x + y - z = 6$
2.  $x + 2y + z = 0$

**Step 1: Write the augmented matrix**

The augmented matrix represents the system of equations:

$\begin{bmatrix} 1 & 1 & -1 & | & 6 \\ 1 & 2 & 1 & | & 0 \end{bmatrix}$

**Step 2: Perform row operations to get the matrix in row-echelon form**

We want to get a leading 1 in the first row, first column (which we already have) and zeros below it.  

* **R2 = R2 - R1** (Replace row 2 with row 2 minus row 1):

$\begin{bmatrix} 1 & 1 & -1 & | & 6 \\ 0 & 1 & 2 & | & -6 \end{bmatrix}$

Now the matrix is in row-echelon form.

**Step 3: Convert the matrix back into equations**

The matrix represents the following equations:

1.  $x + y - z = 6$
2.  $y + 2z = -6$

**Step 4: Use back substitution to solve for the variables**

We have two equations and three unknowns, so we'll solve for x and y in terms of z.

From equation (2):

* $y = -6 - 2z$

Substitute this value of y into equation (1):

* $x + (-6 - 2z) - z = 6$
* $x - 6 - 2z - z = 6$
* $x - 3z = 12$
* $x = 12 + 3z$

**Solution**

The solution to the system is:

* $x = 12 + 3z$
* $y = -6 - 2z$
* $z = z$ (z is a free variable, meaning it can take any value)

## B

**Given system of equations:**

1.  $10x + y = 8$
2.  $x + y = 3$

**Method: Elimination**

* Subtract equation (2) from equation (1):
   $(10x + y) - (x + y) = 8 - 3$
   $9x = 5$
   $x = \frac{5}{9}$

* Substitute $x = \frac{5}{9}$ into equation (2):
   $\frac{5}{9} + y = 3$
   $y = 3 - \frac{5}{9}$
   $y = \frac{27}{9} - \frac{5}{9}$
   $y = \frac{22}{9}$

**Solution:**

$x = \frac{5}{9}$

$y = \frac{22}{9}$ 

or 

$\left(\frac{5}{9}, \frac{22}{9}\right)$

## C
**Given system of equations:**

1.  $x - y + 4z = -3$
2.  $3x + y - z = 8$
3.  $x + y + z = 2$

**Augmented Matrix:**

$\begin{bmatrix} 1 & -1 & 4 & | & -3 \\ 3 & 1 & -1 & | & 8 \\ 1 & 1 & 1 & | & 2 \end{bmatrix}$

**Row Operations:**

* R2 = R2 - 3R1:  $\begin{bmatrix} 1 & -1 & 4 & | & -3 \\ 0 & 4 & -13 & | & 17 \\ 1 & 1 & 1 & | & 2 \end{bmatrix}$
* R3 = R3 - R1: $\begin{bmatrix} 1 & -1 & 4 & | & -3 \\ 0 & 4 & -13 & | & 17 \\ 0 & 2 & -3 & | & 5 \end{bmatrix}$
* R3 = 2R3 - R2: $\begin{bmatrix} 1 & -1 & 4 & | & -3 \\ 0 & 4 & -13 & | & 17 \\ 0 & 0 & 7 & | & -7 \end{bmatrix}$

**Back Substitution:**

* $7z = -7 \implies z = -1$
* $4y - 13(-1) = 17 \implies y = 1$
* $x - 1 + 4(-1) = -3 \implies x = 2$

**Solution:**

* $x = 2$
* $y = 1$
* $z = -1$

or $(2, 1, -1)$
